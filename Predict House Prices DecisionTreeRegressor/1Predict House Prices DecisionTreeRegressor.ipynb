```python
#we first import pandas, pandas is used for data exploration and manipulation
import pandas as pd
```

# load and examine data

```python
#we then read the data into a pandas dataframe, this data is the iowa house prices data from kaggle
#link here https://www.kaggle.com/datasets/nickptaylor/iowa-house-prices, and is available in the "Data" folder of this project
#you will notice that the data is already split into "test" and "train", this is very good, it means we'll not have to split
#our data ourselves into testing and training datasets. This is important because it is not reccomended to test a model on the same
#dataset it is trained it

#we load the train dataset
house_data_train = pd.read_csv("Data/train.csv")
```

```python
#its always a good idea to view a summary of your data before you continue, the describe() method does this
#if you want a detailed explanation of these statistics, check here
#https://medium.com/@soe.a.prime/unlocking-data-secrets-a-beginners-guide-to-25-50-and-75-percentiles-53f0973e57b4
house_data_train.describe()
```

```python
#its also a good idea to look at the available columns in our data frame
house_data_train.columns
```

# select the prediction or output target 

```python
#next, we select the prediction target, this is simply the data we want our model to predict
#in this case, we want to train our model to predict the "SalePrice" so we do the below
#by convention, the prediction target is called "y"

y = house_data_train.SalePrice
```

```python
y
```

# select the features or input values

```python
#next, we choose "features", in this case and generally, the features are the data in our case, the columns which are used
#as input and used to make predictions about the prediction target, sometimes, we will use all the available data as features
#but in this case, we'll select only a few columns as features

#by convention too, features are called X, (capital)
#for now, you can use any columns you like, except the target as features
#lets use our intuition here and try to choose columns we thing actually affect price in real lifeFireplaces

#NOTE: the DecisionTreeRegressor algorithm that we're going to use can only work with numerical feature values,
#so make sure that the columns you're using for features contain just numbers, no strings

house_data_train_features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

#we then use the columns selected to get the actual values and set these to X
X = house_data_train[house_data_train_features]
X
```

# initialise the model and fit the model with data

```python
#now that we have our input and output data, it's time to start fitting our data to our Machine learning algorithm
#fitting data to the algorithm simply means giving it data and having it figure out patterns in the data, when our algorithm 
#has been fitted with data and it has captured patterns enough to be able to predict(or do any other task we want), it effectively 
#becomes a model

#we shall be using the DecisionTreeRegressor algorithm in this project,for this, we'll be using scikit-learn, which is a 
#machine learning library

from sklearn.tree import DecisionTreeRegressor

#initialise the model
house_model = DecisionTreeRegressor(random_state=1)
#don't worry about the "random_state=1" yet, just make sure you use the same number everytime for consistent results

#we then fit the model, basically, train the model
house_model.fit(X,y)
```

```python
#the model has been fitter/trained, we can now use it to make predictions about the saleprice
```

```python
#now watch the magic happen as i predict house price
price_predictions = house_model.predict(X.head())
#the head() method return the first 5 values, so basically, we have predicted price for the first 5 values
price_predictions
```

```python
#the above is predicted price, but let us compare this with our normal price vales
house_data_train.SalePrice.head()
```

```python
#as you can see, our predictions are spot on,you(me) have just predicted house sale prices
#but wait, its too early to celbrate yet. There's no use predicting values we already know,
#we have to test this on the test data
```

# test accuracy of model by comparing prediction against actual values

```python
#we can however test how accurate our prediction model is by using the below
#metrics to test it against the actual scores, the below value is the mean absolute error or MAE
#the MAE tells us by how much, On average, our predictions are off.
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, house_model.predict(X))
```

```python
#as you can see, our test data does not have the SalePrice column
house_data_test.columns
```

```python
#we then have to select the same features for the test data as the train data
house_data_test_features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X_test = house_data_test[house_data_test_features]
X_test
```
